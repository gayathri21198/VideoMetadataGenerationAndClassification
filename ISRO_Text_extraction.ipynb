```python
import moviepy.editor as mp
import speech_recognition as sr
```

```python
directory = "/content/gdrive/MyDrive/Video-Metadata-Classification/ISRO-TextExtraction/Interview"
```

```python
%cd Interview
```

```python
print(directory)
```

```python
for file in os.listdir(directory):
  filename = os.fsdecode(file)
  print("filename : ", filename)
  path = os.path.join(directory, filename)
  print("path - " , path)
  if filename.endswith(".mp4"):
    video = mp.VideoFileClip(path)
    if not os.path.exists('Interview-AudioFile'):
      os.makedirs('Interview-AudioFile')
    name = './Interview-AudioFile/' + str(filename) + '.wav'
    audio_file = video.audio
    audio_file.write_audiofile(name)
    r = sr.Recognizer()
    with sr.AudioFile(name) as source:
      data = r.record(source)
      text = r.recognize_google(data)
      print("\nThe resultant text from video ", filename ,"is: \n")
      print(text, "\n")
      if not os.path.exists('Interview-AudioFile-transcript'):
        os.makedirs('Interview-AudioFile-transcript')
      transcriptPath = './Interview-AudioFile-transcript/'+ str(filename) + '_transcript.txt'
      with open(transcriptPath, "w") as outputFile:
        outputFile.write(text)
```

```python
for file in os.listdir(directory1):
  filename = os.fsdecode(file)
  print("filename : ", filename)
  path = os.path.join(directory1, filename)
  print("path - " , path)

  videoCap = cv2.VideoCapture(path)
  fps = int(videoCap.get(cv2.CAP_PROP_FPS))

  print("filename - " + filename + " frames per second - " ,fps)

  if not os.path.exists('Text-Video-Frames'):
      os.makedirs('Text-Video-Frames')

  currentframe = 0

  while(True):
    ret,frame = videoCap.read()
    print("Read a new frame - " , ret)
    if ret:
      name = './Text-Video-Frames/frame' + str(currentframe) + '.jpg'
      if currentframe%(10*fps) == 0 :
        cv2.imwrite(name, frame)
        print('successfully written 10th frame')
      currentframe += 1
    else:
      break
```

```python
import cv2
from google.colab.patches import cv2_imshow
for file in os.listdir(directory1):
  filename = os.fsdecode(file)
  print("filename : ", filename)
  video_path = os.path.join(directory1, filename)
  print("video_path - " , video_path)

  cap = cv2.VideoCapture(video_path)
  frame_rate = cap.get(cv2.CAP_PROP_FPS)
  frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

  frame_list = []

  while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
      break
    #img = cv2.imread(frame)
    img = cv2.resize(frame, (720, 480))
    hImg, wImg, _ = img.shape
    boxes = pytesseract.image_to_boxes(img)

    for b in boxes.splitlines():
      b = b.split(' ')
      print(b)
      x, y, w, h = int(b[1]), int(b[2]), int(b[3]), int(b[4])
      cv2.rectangle(img, (x, hImg - y), (w, hImg - h), (50, 50, 255), 1)
      cv2.putText(img, b[0], (x, hImg - y + 13), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (50, 205, 50), 1)

    cv2_imshow(img)
    cv2.waitKey(0)

    img1 = cv2.resize(frame, (600, 360))
    texts = pytesseract.image_to_string(img1)
    print(texts)

    if not os.path.exists('Image-Text-transcript'):
      os.makedirs('Image-Text-transcript')
    transcriptPath = './Image-Text-transcript/'+ str(filename) + '_transcript.txt'
    with open(transcriptPath, "w") as outputFile:
      outputFile.write(texts)

    frame_list.append(img)

  cap.release()

  for i, frame in enumerate(frame_list):
    output_path = './Image-Text-transcript/'+ 'frame_{i}' + '.png'
    cv2.imwrite(output_path, frame)
```
